# Prediction using Decision Tree Algorithm

###  

Importing necessary libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

######  

Reading the dataset

```python
iris=pd.read_csv('C:/Users/Rasika Choraghe/Desktop/Iris.csv')
iris
```

######  

```python
iris.info()
```

```python
iris.Species.value_counts()
```

```python
iris['Species_class']=np.where(iris.Species=='Iris-virginica',1,np.where(iris.Species=='Iris-versicolor',2,3))
```

```python
iris.Species_class.value_counts()
```

```python
iris.columns
```

```python
cols=['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']
```

######  

Train-Test Split

```python
from sklearn.model_selection import train_test_split


train_X, test_X, train_y, test_y = train_test_split( iris[cols],
                                                  iris['Species_class'],
                                                  test_size = 0.2,
                                                  random_state = 123 )
```

```python
param_grid = {'max_depth': np.arange(2, 8),
             'max_features': np.arange(2,5)}
```

######  

Importing Decision Tree Classifier and fit the model

```python
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeClassifier, export_graphviz
tree = GridSearchCV(DecisionTreeClassifier(), param_grid, cv = 10,verbose=1,n_jobs=-1)
tree.fit( train_X, train_y )
```

######  

```python
tree.best_score_
```

```python
tree.best_estimator_
```

```python
tree.best_params_
```

```python
train_pred = tree.predict(train_X)
```

```python
test_pred = tree.predict(test_X)
```

######  

Evaluation Metrics

```python
import sklearn.metrics as metrics
print(metrics.classification_report(test_y, test_pred))
```

```python
clf_tree = DecisionTreeClassifier( max_depth = 4, max_features=2)
clf_tree.fit( train_X, train_y )
```

```python
tree_test_pred = pd.DataFrame( { 'actual':  test_y,
                            'predicted': clf_tree.predict( test_X ) } )
```

######  

Prediction

```python
tree_test_pred.sample( n = 10 )
```

######  

Model Accuracy

```python
metrics.accuracy_score( tree_test_pred.actual, tree_test_pred.predicted )
```

```python
tree_cm = metrics.confusion_matrix( tree_test_pred.predicted,
                                 tree_test_pred.actual,
                                 [1,0] )
sns.heatmap(tree_cm, annot=True,
         fmt='.2f',
         xticklabels = ["Yes", "No"] , yticklabels = ["Yes", "No"] )

plt.ylabel('True label')
plt.xlabel('Predicted label')
```

######  

Decision Tree

```python
from sklearn import tree
fn=['sepal length (cm)','sepal width (cm)','petal length (cm)','petal width (cm)']
cn=['setosa', 'versicolor', 'virginica']
fig, axes = plt.subplots(nrows = 1,ncols = 1,figsize = (15,10), dpi=300)
tree.plot_tree(clf_tree,
               feature_names = fn, 
               class_names=cn,
               filled = True);
fig.savefig('imagename.png')
```
